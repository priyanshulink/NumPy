**Deep Copy and Sallow Copy**

```python
"""
Sallow copy-
         we can make another notebook that reference to the same

         arr ----------┐
                       │
                       ▼
                 [10 20 30]
                       ▲
                       │
         copy ---------┘
  if we write something in copy original also change they have same notebook
"""
arr = np.array([10,20,30,40,50])
new_arr = arr.view()
# for sallow copy " .view() " is used
new_arr[0] = 100

print(new_arr) # this is copy array

#original array
print(arr)
```

```python
"""
Deep Copy -
      You make a new notebook.

      Original ------> [Notebook A]

      Copied --------> [Notebook B]

    Changing Notebook B does not affect Notebook A.

  """

arr = np.array([10, 20, 30])
copy_arr = arr.copy()
# for deep copy " .copy() " is used
copy_arr[0] = 100

print("Original:", arr)
print("Copy    :", copy_arr)
```

**Advanced matrix Manipulation**


*   Stacking



```python
# Stacking - combine tow are more array into a singal array
# stacking function

#vartical stacking

a = np.array([1,2,3])
b = np.array([4,5,6])

c = np.vstack((a,b))
print(c)
```

```python
# Horizontal stacking

a = np.array([1,2,3])
b = np.array([4,5,6])

c = np.hstack((a,b))
print(c)
```

```python
# Column Stacking - this convert 1 D array into column
print(np.column_stack((a,b)))
```

```python
# Stacking - unlike hstack, vstack and stack() create new axis
print(np.stack((a,b),axis = 1))
```

**spliting array**

```python
# np.split() - it split into two equal size part (most importent)
arr = np.array([10,20,30,40,50,60])
new_arr = np.split(arr,2)
print(new_arr)
```

```python
# np.array_split() - thy can split unevenly
arr = np.array([10,20,30,40,50,60])
new_arr = np.array_split(arr,3)
print(new_arr)
```

```python
# np.hsplit -
arr = np.array([[10,20,30,50],[40,50,60,70]])
new_arr = np.hsplit(arr,2)
print(new_arr)
```

```python
# np.vsplit
arr = np.array([[10,20,30,50],[40,50,60,70]])
new_arr = np.vsplit(arr,2)
print(new_arr)
```
