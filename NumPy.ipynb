**Random generation function**

```python
# rand() - it generate decimal number between 0 and 1
arr = np.random.rand(5)
print(arr)
```

```python
# randn()  - it generate decimal number between -3 to 3
arr = np.random.randn(5)
print(arr)
```

```python
# randint()  - it generate random integer
arr = np.random.randint(1,5)
print(arr)
arr1 = np.random.randint(1,10,5) #this will give 5 integer between 1 to 10
print(arr1)
```

```python
# random 2D arrary
arr = np.random.randint(1,10,(3,4))
print(arr)
```

```python
# choice()
arr = np.random.choice([10,20,30,40,50]) #this will give any number that is in the list
print(arr)
```

```python
""" seed()

Think of seed() like a bookmark 📖

Imagine a book with thousands of random numbers.

seed(10) → Start reading from page 10.
seed(20) → Start reading from page 20.
seed(100) → Start reading from page 100.

The seed is not the output. It just decides where the random number generator starts
"""

np.random.seed(20)

print(np.random.randint(1, 100))
print(np.random.randint(1, 100))
print(np.random.randint(1, 100))
```

**Array Attribute**


```python
arr = np.array([[10, 20, 30],
                [40, 50, 60]])
print(arr)
```

```python
print(arr.shape) # tell you number of row and col.
```

```python
print(arr.size) # tell you total number of element
```

```python
print(arr.dtype) # tell you data type of element
```

```python
print(arr.ndim) # tell how many dimention array has
```

```python
print(arr.itemsize) # tell you how many byte one element occupies
```

```python
print(arr.nbytes) # tell total memenory used by array
```

```python
print(arr.T) # it swap row and column
```

```python
for i in arr.flat: # flat - is used to print each element of array
  print(i)
```
